```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#carregando dados
df_titanic = pd.read_csv("/content/drive/MyDrive/Cp4/Titanic-Dataset.csv")
df_titanic.head()

# Quantidade de Linhas e Colunas
num_linhas, num_colunas = df_titanic.shape

#Dados Faltantes (Null)
dados_faltantes = df_titanic.isnull().sum()

# Verificar os tipos de dados de cada coluna
tipos_de_dados = df_titanic.dtypes
#colunas que possuem dados categoricos
colunas_categoricas = df_titanic.select_dtypes(include=['object']).columns

# Calcular métricas descritivas para a coluna 'Fare'
fare_moda = df_titanic['Fare'].mode()[0]
fare_media = df_titanic['Fare'].mean()
fare_mediana = df_titanic['Fare'].median()
fare_variancia = df_titanic['Fare'].var()
fare_desvio_padrao = df_titanic['Fare'].std()

# Calcular métricas descritivas para a coluna 'Age'
age_moda = df_titanic['Age'].mode()[0]
age_media = df_titanic['Age'].mean()
age_mediana = df_titanic['Age'].median()
age_variancia = df_titanic['Age'].var()
age_desvio_padrao = df_titanic['Age'].std()



quantidade_embarques = df_titanic['Embarked'].value_counts()


# Proporção de sobreviventes
proporcao_sobreviventes = df_titanic['Survived'].mean()


# Passageiros com a menor e a maior idade
menor_idade = df_titanic['Age'].min()
maior_idade = df_titanic['Age'].max()


# Configuração do estilo dos gráficos
sns.set_theme(style="darkgrid")
```

RELATORIO

```
print("RELATORIO")

#quantidade de linhas e colunas
print(f"\nQuantidade de Linhas: {num_linhas}")
print(f"Quantidade de Colunas: {num_colunas}")

# Verificar os tipos de dados de cada coluna
print("\nTipos de Dados por Coluna:")
print(tipos_de_dados)
#colunas que possuem dados categoricos
print("\nColunas Categóricas Identificadas:")
print(colunas_categoricas)
#dados faltantes por coluna
print("\nDados Faltantes por Coluna:")
print(dados_faltantes)



# Exibir resultados para a coluna 'Fare'
print("\nMétricas Descritivas para a coluna 'Fare':")
print(f"Moda: {fare_moda}")
print(f"Média: {fare_media}")
print(f"Mediana: {fare_mediana}")
print(f"Variância: {fare_variancia}")
print(f"Desvio Padrão: {fare_desvio_padrao}")

# Exibir resultados para a coluna 'Age'
print("\nMétricas Descritivas para a coluna 'Age':")
print(f"Moda: {age_moda}")
print(f"Média: {age_media}")
print(f"Mediana: {age_mediana}")
print(f"Variância: {age_variancia}")
print(f"Desvio Padrão: {age_desvio_padrao}")

#QUANTIDADES DE EMBARQUES POR PORTO
print("\nQuantidade de Embarques por Porto:")
print(quantidade_embarques)

# Proporção de sobreviventes
print("\nProporção de Sobreviventes:")
print(f"{proporcao_sobreviventes:.2%}")

# Passageiros com a menor e a maior idade
print("\nMenor e Maior Idade dos Passageiros:")
print(f"Menor Idade: {menor_idade} anos")
print(f"Maior Idade: {maior_idade} anos")





# 1. Gráfico de Barras: Quantidade de Sobreviventes e Não Sobreviventes por Classe de Embarque
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Pclass', hue='Survived', data=df_titanic, palette='Set2', edgecolor='black')
plt.title('Quantidade de Sobreviventes e Não Sobreviventes por Classe de Embarque', fontsize=16)
plt.xlabel('Classe de Embarque', fontsize=14)
plt.ylabel('Quantidade de Passageiros', fontsize=14)
plt.legend(title='Sobreviveu', labels=['Não', 'Sim'], fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12)
plt.show()




# 2. Gráfico de Barras: Quantidade de Homens e Mulheres
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Sex', data=df_titanic, palette='pastel', edgecolor='black')
plt.title('Quantidade de Homens e Mulheres', fontsize=16)
plt.xlabel('Sexo', fontsize=14)
plt.ylabel('Quantidade', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12)
plt.show()

# 3. Gráfico de Barras: Quantidade de Sobreviventes (0 e 1)
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Survived', data=df_titanic, palette='pastel', edgecolor='black')
plt.title('Quantidade de Sobreviventes (0 = Não Sobreviveu, 1 = Sobreviveu)', fontsize=16)
plt.xlabel('Sobreviveu', fontsize=14)
plt.ylabel('Quantidade', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12)
plt.show()

# 4. Histograma: Frequência de Idade
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic['Age'].dropna(), bins=30, kde=True, color='blue', edgecolor='black')
plt.title('Frequência de Idade dos Passageiros', fontsize=16)
plt.xlabel('Idade', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# 5. Histograma: Frequência de Fare (Valor Pago na Passagem)
plt.figure(figsize=(10, 6))
sns.histplot(df_titanic['Fare'], bins=30, kde=True, color='green', edgecolor='black')
plt.title('Frequência de Valor Pago na Passagem (Fare)', fontsize=16)
plt.xlabel('Valor Pago (Fare)', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# 6. Gráfico de Barras: Frequência de Porto de Embarque
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Embarked', data=df_titanic, palette='pastel', edgecolor='black')
plt.title('Frequência de Porto de Embarque', fontsize=16)
plt.xlabel('Porto de Embarque', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12)
plt.show()
```
